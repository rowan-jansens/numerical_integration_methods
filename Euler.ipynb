```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
#Exploring Euler's method
#By - Rowan Jansens

#loop though differt delta-x values
for dx in np.arange(0.000001, 1, 0.01):
    
    e = 1
    data = []
    domain = []
    #apply Euler's method
    for x in np.arange(0,1,dx):
        data.append(e)
        domain.append(x)
        #update the estimate besed on inital slop times dx
        ddx = e
        e += (ddx * dx)
    #plot it
    plt.plot(domain, data, c = "white", lw = 1)
    
#make two arrays to store x and y values
data = []
domain = []

#fill lists with the true value of e
for i in np.arange(0, 1, 0.00001):
    domain.append(i)
    data.append(np.exp(i))

#finaly plot real exp(x) to compare
plt.rcParams["figure.figsize"] = (13,13)
plt.plot(domain, data, c = "blue", lw =3)

```

```python

#define the ODE here: dy/dx = f(x,y,...)
def ode(x, y):
    return np.sin(x**2) - y

#inital condition
x0 = -4
y0 = 2

#plotting range
minx = 0
maxx = 5

#set step size
step = 0.001

#initaliza data array
plotx = []
ploty = []

#solve ODE backward and forward from inital condition

#first set curret value to inital condition
val = y0
for i in np.arange(x0, minx,-1 * step):
    #apply Euler's method to update value
    val -= (step * ode(i, val))
    #add values to array
    plotx.append(i)
    ploty.append(val)

#repeat for forward dirrection
val = y0
for i in np.arange(x0, maxx, step):
    val += (step * ode(i, val))
    plotx.append(i)
    ploty.append(val)
    
#plot the results
plt.rcParams["figure.figsize"] = (10,9)
plt.scatter(plotx, ploty, c = "purple")

#compare results at https://www.geogebra.org/m/W7dAdgqc
```
