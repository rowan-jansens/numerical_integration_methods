```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
#define the ODE here: dy/dx = f(x,y,...)
def ode(x, y):
    return y

#set step size
step = 0.01

#inital condition
val = 1
plotx = []
ploty = []

for t in np.arange(0, 1 + step, step):
    plotx.append(t)
    ploty.append(val)
    
    kl = ode(t, val)
    km = ode((t + step / 2), (val + step / 2 * kl))
    kc = ode((t + step / 2), (val + step / 2 * km))
    kr = ode((t + step), (val + step * kc))
    val += step * (kl + 2*km + 2*kc + kr) / 6
    
    
#plot the results
plt.rcParams["figure.figsize"] = (10,9)
plt.scatter(plotx, ploty, c = "purple", lw = 1)

#fill lists with the true value of e
plotx = []
ploty = []
for i in np.arange(0, 1.01, 0.01):
    plotx.append(i)
    ploty.append(np.exp(i))

#finaly plot real exp(x) to compare
plt.plot(plotx, ploty, c = "white", lw = 1)
```

```python
#define the ODE here: dy/dx = f(x,y,...)
def ode(x, y):
    return np.sin(x**2) - y

#inital condition
x0 = -4
y0 = 2

#plotting range
minx = 0
maxx = 5

#set step size
step = 0.01

#initaliza data array
plotx = []
ploty = []

#solve ODE backward and forward from inital condition

#first set curret value to inital condition
val = y0
for t in np.arange(x0, minx,-1 * step):
    plotx.append(t)
    ploty.append(val)
    
    kl = ode(t, val)
    km = ode((t + step / 2), (val + step / 2 * kl))
    kc = ode((t + step / 2), (val + step / 2 * km))
    kr = ode((t + step), (val + step * kc))
    val -= step * (kl + 2*km + 2*kc + kr) / 6

#repeat for forward dirrection
val = y0
for t in np.arange(x0, maxx, step):
    plotx.append(t)
    ploty.append(val)
    
    kl = ode(t, val)
    km = ode((t + step / 2), (val + step / 2 * kl))
    kc = ode((t + step / 2), (val + step / 2 * km))
    kr = ode((t + step), (val + step * kc))
    val += step * (kl + 2*km + 2*kc + kr) / 6
    
    
#plot the results
plt.rcParams["figure.figsize"] = (10,9)
plt.scatter(plotx, ploty, c = "purple")

#compare results at https://www.geogebra.org/m/W7dAdgqc
```
